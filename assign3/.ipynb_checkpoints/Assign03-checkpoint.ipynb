```python
#爬全部
url="https://www.yelp.com/search?find_desc=mexican+food&find_loc=Washington%2C+DC"
while url!='':
    r = requests.get(url)
    data = r.text    
    soup = bs(data, "html.parser")
    myList = []
    lists=soup.findAll('li',{'class':'regular-search-result'})
    for list in lists:
        restaurants = list.findAll('div',{'class':'search-result natural-search-result'})
        for restaurant in restaurants:
            c = []
            name = restaurant.find('a',{'data-analytics-label':'biz-name'})
            c.append(name.getText().strip())
            address = restaurant.find('address').getText().strip()
            address2 = restaurant.find('address').find('br').getText().strip()
            StreetAddress = address[0:len(address)-len(address2)]
            City=address2.split(',')
            c.append(StreetAddress)
            c.append(City[0])
            State=City[1]
            State=State.split(' ')
            c.append(State[1])
            c.append(State[2])
            phone = restaurant.find('span',{'class':'biz-phone'})
            c.append(phone.getText().strip())
            reviewNum = restaurant.find('span',{'class':'review-count rating-qualifier'}).getText().strip()
            reviewNum = reviewNum.split(' ')
            c.append(int(reviewNum[0]))
            rating = restaurant.find('div',{'class':'rating-large'})
            rating = rating.img.get('alt')
            rating = rating.split(' ')
            c.append(float(rating[0]))
            price = restaurant.find('span',{'class':'bullet-after'}).getText().strip()
            if price == '$':
                price = 10
            if price == '$$':
                price = 20
            if price == '$$$':
                price = 30
            c.append(price)
            myList.append(c)
    nexturl = soup.find('a',{'class':'u-decoration-none next pagination-links_anchor'}).get('href')
    nexturl="https://www.yelp.com"+nexturl
    url = nexturl
print(myList)
```

```python

import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
from matplotlib.pyplot import savefig
def step3_1(fileName,pdfName):


    DF = pd.read_csv(fileName)
    DF = DF[pd.notnull(DF['Price Ranges'])]
    DF = DF[pd.notnull(DF['Rating'])]
    DF = DF.reset_index(drop=True)
   
    df = DF[[8,9]]
    df.columns=['Rating',"Price Range"]
    
    x1=df[(df["Price Range"]==10)]
    x2=df[(df["Price Range"]==20)]
    x3=df[(df["Price Range"]==30)]
    x4=df[(df["Price Range"]==40)]
    
    data_to_plot = [x1, x2, x3, x4]
    print(data_to_plot)
    plt.boxplot(data_to_plot)
    
    plt.title("The Relationship between Rating and Price Range: " + fileName.split(".")[0])
    plt.xlabel("Price Range (number of '$')")
    plt.ylabel("Rating")
    plt.ylim([0.5,5.5])
    savefig(pdfName)
    
    ########
    plt.close()    
```

```python
def handlePrice (df):
    price=[]
    for i in df["Price Range"]:
        if i == "$":
            price.append(10)
        if i == "$$":
            price.append(20)
        if i == "$$$":
            price.append(30)
        if i == "$$$$":
            price.append(40)
        if i == "$$$$$":
            price.append(50)
    rating = [i for i in df['Rating']]
    reviews = [i for i in df['Number of Reviews']]
    df = pd.DataFrame({"Rating":rating,"Price Ranges":price,"Number of Reviews":reviews})
            
```
